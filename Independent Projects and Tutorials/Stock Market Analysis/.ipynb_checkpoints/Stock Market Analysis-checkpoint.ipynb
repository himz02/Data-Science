## Observe moving averages

```python
ma_days = [10, 20, 50]
for ma in ma_days:
    column = 'MA for %s days' %(str(ma))
    
    AAPL[column] = AAPL['Adj Close'].rolling(window = ma).mean()
```

```python
AAPL[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(subplots = False, legend = True, figsize=(10, 4))
```

```python
AAPL['Daily Return'] = AAPL['Adj Close'].pct_change()

AAPL['Daily Return'].plot(figsize = (10, 4), legend = True, linestyle = '--', marker = 'o')
```

### Compare different companies

```python
closing_df = data.DataReader(tech_list, 'yahoo', start, end)['Adj Close']
closing_df
```

```python
tech_rets = closing_df.pct_change()
```

```python
sns.jointplot(x = 'MSFT', y = 'GOOG', data = tech_rets, color = 'seagreen', kind = 'scatter')
```

```python
sns.pairplot(tech_rets.dropna())
```

### How to edit the pair plots?

#### Daily Return pair plot

```python
return_fig = sns.PairGrid(tech_rets)
return_fig.map_upper(plt.scatter, color = 'purple')
return_fig.map_lower(sns.kdeplot, color = 'darkred')
return_fig.map_diag(plt.hist, bins = 30)
```

#### Closing value pair plot

```python
return_fig = sns.PairGrid(closing_df)
return_fig.map_upper(plt.scatter, color = 'purple')
return_fig.map_lower(sns.kdeplot, color = 'darkred')
return_fig.map_diag(plt.hist, bins = 30)
```

### Correlation plots

```python
corr = tech_rets.dropna().corr()
sns.heatmap(corr, cmap = 'Greens', annot = True)
```

```python
corr = closing_df.dropna().corr()
sns.heatmap(corr, cmap = 'Greens', annot = True)
```
